```python
import pandas as pd

df = pd.read_csv("../data/insurance_data.csv")

df.head()
df.shape
df.info()
df.describe()
df.isnull().sum()
df["LossRatio"] = df["TotalClaims"] / df["TotalPremium"]

df["Margin"] = df["TotalPremium"] - df["TotalClaims"]
import matplotlib.pyplot as plt

df["TotalClaims"].hist(bins=30)
plt.show()
df["Province"].value_counts().plot(kind="bar")
plt.show()
plt.scatter(df["TotalPremium"], df["TotalClaims"])
plt.xlabel("Premium")
plt.ylabel("Claims")
plt.show()
import seaborn as sns

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
province_risk = df.groupby("Province")["LossRatio"].mean()

province_risk.plot(kind="bar")
plt.show()
sns.boxplot(x=df["TotalClaims"])
plt.show()
overall_loss_ratio = df["TotalClaims"].sum() / df["TotalPremium"].sum()

print(overall_loss_ratio)
df.groupby("Gender")["LossRatio"].mean()
df.groupby("VehicleType")["TotalClaims"].mean().sort_values(ascending=False)
df.groupby("VehicleType")["TotalClaims"].mean().sort_values(ascending=False)


```
