```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

sys.path.append('../')
from src.data_loader import InsuranceDataLoader
from src.eda_utils import InsuranceEDA

print("Libraries loaded successfully!")
```

```python
# Load data
loader = InsuranceDataLoader(data_path='../data/insurance_data.csv')
df_raw = loader.load_data()
df = loader.clean_data(df_raw)

print(f"Data shape: {df.shape}")
print(f"Columns: {list(df.columns)[:10]}...")
```

```python
# Initialize EDA utilities
eda = InsuranceEDA(df)

# Calculate portfolio metrics
metrics = eda.calculate_portfolio_metrics()

print(f"Total Premium: R{metrics['total_premium']:,.2f}")
print(f"Total Claims: R{metrics['total_claims']:,.2f}")
print(f"Overall Loss Ratio: {metrics['overall_loss_ratio']:.4f} ({metrics['overall_loss_ratio']*100:.2f}%)")
print(f"Average Margin per Policy: R{metrics['avg_margin']:,.2f}")
print(f"Total Policies: {metrics['total_policies']:,}")
print(f"Claim Frequency: {metrics['claim_frequency']:.2%}")
```

```python
# Analyze loss by province
loss_by_province = eda.plot_loss_ratio_by_category('Province')
print(loss_by_province)
```
